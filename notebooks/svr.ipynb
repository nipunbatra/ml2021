```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(0, 50, 51)
```

```python
y = X*4 + 10*np.random.randn(len(X))
```

```python
plt.scatter(X, y, s=50, edgecolors='gray',facecolors='none')
plt.plot(X, X*4)
```

```python
from sklearn.svm import SVR
from ipywidgets import FloatSlider, interact
```

```python
@interact(eps=[1e-3, 1e-2,5, 10,20, 50,100])
def svr_eps(eps):
    plt.scatter(X, y, s=50, edgecolors='gray',facecolors='none')
    plt.plot(X, X*4, label='True line')
    sv = SVR(kernel='linear', degree=1,  epsilon=eps)
    sv.fit(X.reshape(-1, 1), y,)
    plt.plot(X, X*sv.coef_[0] + sv.intercept_, label='EPS = {}'.format(eps))
    plt.legend()
```
