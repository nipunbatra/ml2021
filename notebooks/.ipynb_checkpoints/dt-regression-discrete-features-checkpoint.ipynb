### Gain(S, Outlook)

```python
dfs = df['Minutes Played'].std()
stds = {}
```

```python
variable = 'Wind'
stds[variable] = 0.
for g in df.Outlook.unique():
    df_g = df[df[variable]==g]
    stds[variable]+= df_g['Minutes Played'].std()*(len(df_g)/len(df))
```
