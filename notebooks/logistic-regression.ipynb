```python
import pandas as pd
import numpy as np
```

```python
data = np.array([[0, 0], [0.1, 0], [0.2, 0], [0.7, 1], [0.8, 1]])
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
```

```python
logr = LogisticRegression()
```

```python
logr.fit(data[:, 0].reshape(-1, 1), data[:, 1])
```

```python
logr.coef_
```

```python
logr.intercept_
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
%matplotlib inline
```

```python
plt.scatter(data[:, 0], data[:, 1], c=data[:, 1], cmap=cm.viridis)
plt.axvline(-logr.intercept_[0]/logr.coef_[0],label='Decision Boundary',linestyle='--',color='k',lw=5)
plt.legend()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
x_dum = np.linspace(-5, 5, 100)
plt.plot(x_dum, sigmoid(x_dum*logr.coef_[0] + logr.intercept_[0]))
plt.scatter(data[:, 0], data[:, 1], c=data[:, 1], cmap=cm.viridis)
plt.axvline(-logr.intercept_[0]/logr.coef_[0], lw=2)
plt.axhline(0.5, linestyle='--',color='k',lw=5)
plt.ylabel("P(y=1)")
plt.title("Logistic Regression")
```

```python
lr = LinearRegression()
lr.fit(data[:, 0].reshape(-1, 1), data[:, 1])
plt.plot(x_dum, x_dum*lr.coef_[0]+lr.intercept_)
plt.scatter(data[:, 0], data[:, 1], c=data[:, 1], cmap=cm.viridis)
plt.axhline(0.5, linestyle='--',color='k')
plt.ylabel("P(y=1)")
plt.title("Linear Regression")
```
