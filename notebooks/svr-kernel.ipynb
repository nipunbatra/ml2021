```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(0, 50, 51)
```

```python
y = np.sin(X) + 0.5*np.random.randn(len(X))
```

```python
plt.scatter(X, y, s=50, edgecolors='gray',facecolors='none')
plt.plot(X, np.sin(X))
```

```python
from sklearn.svm import SVR
from ipywidgets import FloatSlider, interact
```

```python
sv = SVR(kernel='rbf',  epsilon=0.2, C=100)
```

```python
sv.fit(X.reshape(-1, 1), y)
```

```python
plt.scatter(X, y, s=50, edgecolors='gray',facecolors='none')
plt.plot(X, np.sin(X), label='True line')
plt.plot(X, sv.predict(X.reshape(-1, 1)))
```
